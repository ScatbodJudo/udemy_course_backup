```python
import pandas
import urllib.request
import json
import folium
```

```python
#all earthquakes from the past hour:
#url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_hour.geojson"

#all earthquakes from the past day:
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.geojson"

#to eliminate the certificate error, run the "install Certificates.command" from the python folder
#or see https://stackoverflow.com/questions/35569042/python-3-ssl-certificate-verify-failed

#open and close connection to url and read html
with urllib.request.urlopen(url) as response:
    html = response.read()
    print(response.getcode())
```

```python
#load data form geoJson
#throw data into a list object (is this neesesary?)
data = json.loads(html)
#for i in data["features"]: print( i["properties"]["place"])
quakes = []
mags = []
#make a list of locations from the json data
for i in data["features"]: 
    quakes.append(i["geometry"]["coordinates"])
#make a list of magnitues from the json data    
for i in data["features"]:
    mags.append(i["properties"]["mag"])

#throw earthquake locations in a pandas dataframe
df_quakes = pandas.DataFrame(quakes)

#rename the ugly columns. (i don't know what that third attribute is...)
df_quakes.columns =['lon','lat','depth']

#append magnitude list to the quakes data frame
df_quakes["mag"]=mags

#proportional symbol range: 100-200
df_quakes['symbol_size'] = df_quakes.mag.apply(lambda x: x*5)


def opacityNormalizer(depth):
    depth_range = df_quakes['depth'].max()-df_quakes['depth'].min()
    opacity = 1-(depth/depth_range*.66)
    return opacity




df_quakes
```

```python
df_quakes.mean()
#calculate mean center
mean_lat = df_quakes.mean().lat
mean_lon = df_quakes.mean().lon
```

```python
#center = 

#make a map!
m = folium.Map(location=[mean_lat,mean_lon],zoom_start=2.5)
#put some shit on the map!

#make a feature group
epicenters = folium.FeatureGroup(name="earthquakes")


#iterate over data and plot them all at once
#use zip to iterate through data and make circles in map
#add a proportional symbol functionality to earthquake symbols
for lat,lon, mag, symbol_size,depth in zip(df_quakes.lat, df_quakes.lon, df_quakes.mag, df_quakes.symbol_size,df_quakes.depth):
    folium.CircleMarker(
        radius=symbol_size,
        location=[lat, lon],
        popup= folium.Popup("Magnitude: "+str(mag)+", Depth: "+str(depth)+"km",parse_html=True),
        color='crimson',
        opacity = opacityNormalizer(depth),
        fill=False,
    ).add_to(epicenters)
#after adding the data to the 'epicenters' feature group, you have to add the features to the map
epicenters.add_to(m)

m
```
