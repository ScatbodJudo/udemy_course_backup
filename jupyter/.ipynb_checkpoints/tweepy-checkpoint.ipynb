```python
tweets_filename = "tweet_stream.txt"
tweets_file = open(tweets_filename,"r+")

track_words = ['love','peace','trump']

class Listener(StreamListener):
    def on_data(self, data):
        tweets_file.write(data)
        print(data)
        return True
    
    def on_error(self, status):
        print(status)

if __name__ == '__main__':
    
    listener = Listener()
    auth = OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
    auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
    stream = Stream(auth, listener)
    
    stream.filter(track=track_words)
```

```python
#list object to hold each line of the json file
tweets = []

tweets_data = {
    "text":[],
    "lang":[],
    "country":[],
    "location":[]
}

tweets_file = open(tweets_filename,"r")
for line in tweets_file:
    try:
        tweet = json.loads(line.strip())
        tweets.append(tweet)
        tweets_data["text"].append(tweet["text"])
        tweets_data["lang"].append(tweet["lang"])
        if tweet["place"] != None: tweets_data["country"].append(tweet["place"]["country"])
        else: tweets_data["country"].append(None)
        if tweet['user']['location'] != None: tweets_data["location"].append(tweet['user']['location'])
        else: tweets_data['location'].append(None)
    except:
        continue
print("number of tweets: "+ str(len(tweets)))

df_tweets = pandas.DataFrame.from_dict(tweets_data)

#load json file data ("tweet") into pandas dataframe
#as per tutorial, but I like to use .from_dict
#df_tweets["text"] = map(lambda tweet: tweet["text"], tweets)
#df_tweets["lang"] = map(lambda tweet: tweet["lang"], tweets)
#df_tweets["country"] = map(lambda tweet: tweet["place"]["country"] if tweet['place'] != None else None, tweets)

tweets_file.close()
df_tweets
```
